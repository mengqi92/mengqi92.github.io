```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from ipywidgets import interact
import seaborn as sns

sns.set_context('poster')
# sns.set_style('whitegrid')
```

```python
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei'] #指定默认字体  
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
x0 = np.array([600000, 400000])
A = np.array([[0.95, 0.03], [0.05, 0.97]])
```

```python
def markov(x0):
    xi = x0
    while True:
        yield np.dot(A, xi)
        xi = np.dot(A, xi)

m = markov(x0)
markov_1000 = np.array([next(m) for _ in range(1000)])
```

```python
plt.scatter(markov_1000[:100][:, 0], markov_1000[:100][:, 1])
```

```python
@interact
def func(idx=(0, 1000)):
    plt.figure()
    plt.title('city: {:.2f}, rural: {:.2f}'.format(*markov_1000[idx]))
    plt.ylim(300000, 700000)
    plt.xlim(300000, 700000)
    plt.scatter(markov_1000[:idx][:, 0], markov_1000[:idx][:, 1])
    plt.scatter(*markov_1000[idx-1], c='yellow', s=60, marker='o')
    plt.scatter(*markov_1000[idx], c='red', s=60, marker='o')
```

```python
eigvalues, eigvectors = np.linalg.eig(A)
```
